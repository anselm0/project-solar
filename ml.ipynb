```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import sklearn
```

```python
zipsolar_url = "data/texas_solar_data.csv"
zipincome_url = "data/median_income_texas_zip.csv"

sdf = pd.DataFrame.from_csv(zipsolar_url)
idf = pd.DataFrame.from_csv(zipincome_url, )
```

```python
sdf = sdf.reset_index()
sdf
```

```python
idf = idf.reset_index()
idf.head()
```

```python
idf = idf.rename(columns = {'Zip Code':'region_name'})
idf.head()
```

```python
df = pd.merge(sdf, idf, how='left', on='region_name')
```

```python
#create a new column with existing installs/count qualified for a percentage
df = df.assign(existing_installs_percent=lambda df: (df.existing_installs_count/df.count_qualified)*100)
```

```python
df.head()
```

```python
# Prepare the data
X = np.c_[df["Avg. Income/H/hold"]]
y = np.c_[df["existing_installs_percent"]]
```

```python
# Visualize the data
df.plot(kind='scatter', x="Avg. Income/H/hold", y="existing_installs_percent")
plt.show()
```

```python
# Select a linear model
lin_reg_model = sklearn.linear_model.LinearRegression()
```

```python
# Train the model
lin_reg_model.fit(X, y)
```

```python
# Make a prediction for Cyprus
X_new = [[22587]] # Cyprus' GDP per capita
print(lin_reg_model.predict(X_new)) # outputs [[ 5.96242338]]
```
